````python
class Model:
    pass

from dataclasses import dataclass

@dataclass
class Image:
    url: str
    height: int | None
    width: int | None

from typing import Any
ResponseDict = dict[str, Any]
````

````python
from pprint import pformat, pprint

class User(Model):
    def __init__(self, d: ResponseDict) -> None:
        self._display_name = d["display_name"]
        self._spotify = d["external_urls"]["spotify"]
        self._href = d["href"]
        self._images = [Image(i["url"], i["height"], i["width"]) for i in d["images"]]
        self._followers = d["followers"]["total"]
        self._id = d["id"]
        self._uri = d["uri"]
        # save reference to original ResponseDict
        self._source = d
    
    @property
    def display_name(self) -> str:
        return self._display_name

    @property
    def spotify(self) -> str:
        return self._spotify
    
    @property
    def followers(self) -> int:
        return self._followers
    
    @property
    def id(self) -> str:
        return self._id
    
    @property
    def uri(self) -> str:
        return self._uri

    def __repr__(self) -> str:
        dict_str = pformat(self._source, indent=2)
        dict_str = " " + dict_str.removeprefix("{")  # get rid of opening {
        dict_str = dict_str.removesuffix("}")  # get rid of closing }
        return f"User({{\n{dict_str}\n}})"
    
    def __str__(self) -> str:
        return f"<User [{self._display_name}]>"

    def get_top_tracks(self) -> list[str]:
        endpoint = "/me/top/tracks"
        url = URL.format(endpoint=endpoint)
        headers = {
            "Authorization": f"Bearer {access_token}",
            "Content-Type": "application/json"
        }
        response = requests.get(url, headers=headers)
        # print(response.json() if response.ok else response.reason)
        d = response.json()
        return [track_item["name"] for track_item in d["items"]]
    
````
